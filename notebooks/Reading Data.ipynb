**Importing requierd librarries**

```python
import os
import pandas  as pd
import numpy as np
```

**Data Frames generation and reading data**

```python
main_path = 'C:/Users/Maedeh/Desktop/TenisMatchPro/data/processed'
# Defining the empty data frames
MatchAwayTeamInfo = pd.DataFrame()
MatchEventInfo = pd.DataFrame()
MatchVotesInfo = pd.DataFrame()
MatchVenueInfo = pd.DataFrame()
MatchTournamentInfo = pd.DataFrame()
MatchSeasonInfo = pd.DataFrame()
MatchRoundInfo = pd.DataFrame()
MatchAwayScoreInfo = pd.DataFrame()
MatchHomeTeamInfo = pd.DataFrame()
MatchHomeScoreInfo = pd.DataFrame()
GameInfo = pd.DataFrame()
PowerInfo = pd.DataFrame()
OddsInfo = pd.DataFrame()
PeriodInfo = pd.DataFrame()
MatchTimeInfo = pd.DataFrame()

date_folder_names=os.listdir(main_path)
# This for is for opening the date folders
for dfn in date_folder_names : 
        sec_path = f'{main_path}/{dfn}/data/raw'    
        folder_names = os.listdir(sec_path)
        # This for is for openning folder names in each date folder 
        for fn in folder_names :
            thrd_path = f'{sec_path}/{fn}' 
            parquet_names = os.listdir(thrd_path)
            # This for is for reading parquet files in each folder 
            for pn in parquet_names : 
                directory_path = os.path.join(thrd_path, pn)
                # MatchAwayTeamInfo Data Frame generation
                if pn.startswith('away_team_1') :
                     MatchAwayTeamInfo = pd.concat ([MatchAwayTeamInfo, pd.read_parquet(directory_path)])
                # MatchEventInfo Data Frame generation
                if pn.startswith('event') :
                     MatchEventInfo = pd.concat ([MatchEventInfo, pd.read_parquet(directory_path)])
               # MatchVotesInfo Data Frame generation
                if pn.startswith('votes') :
                     MatchVotesInfo = pd.concat ([MatchVotesInfo, pd.read_parquet(directory_path)])
                # MatchVenueInfo Data Frame generation
                if pn.startswith('venue') :
                     MatchVenueInfo = pd.concat ([MatchVenueInfo, pd.read_parquet(directory_path)])
               #MatchTournamentInfo Data Frame generation
                if pn.startswith('tournament') :
                     MatchTournamentInfo = pd.concat ([MatchTournamentInfo, pd.read_parquet(directory_path)])
               #MatchSeasonInfo Data Frame generation
                if pn.startswith('season') :
                     MatchSeasonInfo = pd.concat ([MatchSeasonInfo, pd.read_parquet(directory_path)])
                # MatchRoundInfo Data Frame generation
                if pn.startswith('round') :
                     MatchRoundInfo = pd.concat ([MatchRoundInfo, pd.read_parquet(directory_path)])
               # MatchAwayScoreInfo Data Frame generation
                if pn.startswith('away_team_score') :
                     MatchAwayScoreInfo = pd.concat ([MatchAwayScoreInfo, pd.read_parquet(directory_path)])
               # MatchHomeTeamInfo Data Frame generation
                if pn.startswith('home_team_1') :
                     MatchHomeTeamInfo = pd.concat ([MatchHomeTeamInfo, pd.read_parquet(directory_path)])
                # MatchHomeScoreInfo Data Frame generation
                if pn.startswith('home_team_scor') :
                     MatchHomeScoreInfo = pd.concat ([MatchHomeScoreInfo, pd.read_parquet(directory_path)])
               #GameInfo Data Frame generation
                if pn.startswith('pbp') :
                     GameInfo = pd.concat ([GameInfo, pd.read_parquet(directory_path)])
               #PowerInfo Data Frame generation
                if pn.startswith('power') :
                     PowerInfo = pd.concat ([PowerInfo, pd.read_parquet(directory_path)])
                #OddsInfo Data Frame generation
                if pn.startswith('odd') :
                     OddsInfo = pd.concat ([OddsInfo, pd.read_parquet(directory_path)])
               #PeriodInfo Data Frame generation
                if pn.startswith('stat') :
                     PeriodInfo = pd.concat ([PeriodInfo, pd.read_parquet(directory_path)])
               #MatchTimeInfo Data Frame generation
                if pn.startswith('time') :
                     MatchTimeInfo = pd.concat ([MatchTimeInfo, pd.read_parquet(directory_path)])
               
```

Saving data in excel files for rereading in other notebooks

```python
main2_path = 'C:/Users/Maedeh/Desktop/TenisMatchPro/data/external'
MatchAwayTeamInfo.to_csv(f'{main2_path}/MatchAwayTeamInfo.csv' , index= False)
MatchEventInfo.to_csv(f'{main2_path}/MatchEventInfo.csv' , index= False)
MatchVotesInfo.to_csv(f'{main2_path}/MatchVotesInfo.csv' , index= False)
MatchVenueInfo.to_csv(f'{main2_path}/MatchVenueInfo.csv' , index= False)
MatchTournamentInfo.to_csv(f'{main2_path}/MatchTournamentInfo.csv' , index= False)
MatchSeasonInfo.to_csv(f'{main2_path}/MatchSeasonInfo.csv' , index= False)
MatchRoundInfo.to_csv(f'{main2_path}/MatchRoundInfo.csv' , index= False)
MatchAwayScoreInfo.to_csv(f'{main2_path}/MatchAwayScoreInfo.csv' , index= False)
MatchHomeTeamInfo.to_csv(f'{main2_path}/MatchHomeTeamInfo.csv' , index= False)
MatchHomeScoreInfo.to_csv(f'{main2_path}/MatchHomeScoreInfo.csv' , index= False)
GameInfo.to_csv(f'{main2_path}/GameInfo.csv' , index= False)
PowerInfo.to_csv(f'{main2_path}/PowerInfo.csv' , index= False)
OddsInfo.to_csv(f'{main2_path}/OddsInfo.csv' , index= False)
PeriodInfo.to_csv(f'{main2_path}/PeriodInfo.csv' , index= False)
MatchTimeInfo.to_csv(f'{main2_path}/MatchTimeInfo.csv' , index= False)
```
